### DESCRIPTIVE ANALYTICS FOR NUMERICAL COLUMNS

```python
num_col=[i for i in df.columns if df[i].dtype!='O'] # creating a list of numerical columns
# lists all the numerical columns in the dataset
```

```python
num_col # shows the numerical columns
```

```python
# calculates and prints the descriptive statistics
for col in num_col:
  print(f"Column: {col}")
  print(f"  Mean: {df[col].mean()}") # calculates and prints the mean of the column
  print(f"  Median: {df[col].median()}") #calculates and prints the median of the column
  print(f"  Mode: {df[col].mode()[0]}")  # Mode can have multiple values, so we take first one as mode, so we use [0] in the code
  print(f"  Standard Deviation: {df[col].std()}") # calculates and prints the SD of the column
  print("\n")
```

Mean-- Mean calculates the average value of the data points in the columns.It is sensitive to outliers.

Median-- Median is the middle value when it is sorted in the ascending order.It is less sensitive to outliers than compared to the mean.

Mode-- The most frequent value in the column.As the data have more than one mode we used [0] so that we choose the first frequently
occured value.

Standard Deviation-- It measures how the data is spreded out.If the data has higher SD then the data points are widely spread over wide range otherwise it is clustered closely around the mean.  

 As we see there is a significant differnce between the Mean and Median in the columns.So there is a chance of Outliers.

### STANDARDIZATION OF NUMERICAL VALUES

Standardization also known as Z-score normalization, is a technique used to rescale features in the dataset.It transforms the data into a standard normal distribution with a mean of 0 and SD of 1.

Standardization can reduce the impact of outliers, as it centers the data around the mean.

The formula for calculating the z-score of a value X in the dataset is:


Z=
X-μ/σ
​

Where:


Z is the z-score.

X is the original value.

μ is the mean of the feature values.

σ is the standard deviation of the feature values.

```python
df.columns
```

```python
num_col
```

```python
for col in num_col:
  print(f'Mean of {col}:{np.mean(df[col])}')
  print(f'Standard Deviation of {col}:{np.std(df[col])}')
  print("\n")
```

```python
## z=x-mean/sd
```

```python
data=pd.DataFrame()
for col in num_col:
    data[col + '_zscore'] = (df[col] - np.mean(df[col])) / np.std(df[col])
print(data.head())
```

### CONVERSION OF CATEGORICAL DATA INTO DUMMY VARIABLES

One-hot Encoding is a technique used to convert the categorical column into a numerical format that machine learning models can understand.

The One-hot Encoding is used to apply for the nominal characters.

Many algorithms only work with numbers. So we use One-hot encoding to turn the categorical columns into numbers.

It creates new binary columns for each category in a categorical feature. Each new column represents one category, and a value of 1 in that column indicates the presence of that category for a particular data point, while 0 indicates its absence.

IMPORTANCE OF DATA PREPROCESSING STEPS



1.   Standardization:
*    Standardization transforms numerical data so that it has a mean of 0 and a standard deviation of 1. This is important for machine learning models that are sensitive to feature scaling.

*    Without standardization, models such as KNN, SVM, or neural networks may produce biased results, as features with larger values (e.g., sales value) would dominate features with smaller ranges (e.g., discount rate). Standardization ensures that all features contribute equally, improving the model's performance and training efficiency.

2.   One Hot Encoding:
*   Categorical variables must be converted into a numerical format for machine learning models to interpret them correctly. One-hot encoding transforms each category into a binary column, ensuring that the model doesn't misinterpret categorical data as ordinal.


*   By applying one-hot encoding to columns like City, Brand, and SKU, models can make better use of categorical information without implying any order. This transformation is critical for algorithms such as linear regression, decision trees, and random forests to work accurately with categorical data.
