```python
# Split the single column into lists of items
df['Transactions'] = df.iloc[:, 0].str.split(',')
```

```python
# Verify the first few rows of the 'Transactions' column
df['Transactions'].head()
```

```python
# Convert the list of transactions into a one-hot encoded dataframe
transactions = df['Transactions'].tolist()
te = TransactionEncoder()
te_array = te.fit(transactions).transform(transactions)
encoded_df = pd.DataFrame(te_array, columns=te.columns_)
```

```python
encoded_df
```

```python
# Applying apriori algorithm
frequent_itemsets = apriori(encoded_df,min_support = 0.05,use_colnames=True)
```

```python
frequent_itemsets
```

```python
# Generate association rules
association_rules(frequent_itemsets, metric="confidence", min_threshold=0.001,num_itemsets=len(frequent_itemsets))
```

**Interview Questions**

1. What is lift and why is it important in Association rules?

Lift is a measure of how much more likely the consequent (item Y) is purchased when the antecedent (item X) is purchased, compared to its usual purchase rate. In other words, it quantifies the strength of association between two items.

Lift > 1: Indicates a positive association between the antecedent and consequent, meaning they are more likely to occur together than by chance.

Lift = 1: Implies no association; the items occur together as often as expected by random chance.

Lift < 1: Suggests a negative association, meaning the antecedent and consequent are less likely to occur together.

Importance:
* Identifying true relationships
* Prioritizing rules
* Avoiding misleading insights

2. What is support and Confidence. How do you calculate them?


**Support:** The proportion of transactions that contain both the antecedent (item X) and the consequent (item Y). It measures the frequency of the itemset (X and Y together) in the dataset.

Support(X -> Y) = (Number of transactions containing both X and Y) / (Total number of transactions)


**Confidence:** The probability of purchasing the consequent (item Y) given that the antecedent (item X) has been purchased. It measures the strength of the implication.

Confidence(X -> Y) = (Number of transactions containing both X and Y) / (Number of transactions containing X)
